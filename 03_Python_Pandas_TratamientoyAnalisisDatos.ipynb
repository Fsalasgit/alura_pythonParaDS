# Creando nuevas variables

```python
datos = pd.read_csv('alquiler_residencial.csv', sep =';')
datos.head(10)
```

```python
# Obtener le valor bruto (Osea la suma del valor, mantenimeinto, impuesto)
datos['Valor_bruto'] = (datos['Valor']) + (datos['Mantenimiento']) + (datos['Impuesto'])
datos.head(10)
```

```python
#Obtener el valor por metro cuadrado
datos['Valor_m2'] = (datos['Valor']) / (datos['Area'])
datos.head(10)
```

```python
datos['Valor_m2'] = datos['Valor_m2'].round(2)
datos.head(10)
```

```python
datos['Valor_bruto_m2'] = ((datos['Valor_bruto']) / (datos['Area'])).round(2)
datos.head(10)
```

Identificado y Excluyendo Outliers

```python
datos.boxplot(['Valor'], by =['Tipo'])
```

```python
grupo_tipo = datos.groupby('Tipo')['Valor']
grupo_tipo.groups
```

```python
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
limite_inferior
```

```python
limite_superior
```

```python
datos_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
  eh_tipo = datos['Tipo'] == tipo
  eh_limite = (datos['Valor'] >= limite_inferior[tipo]) & (datos['Valor'] <= limite_superior[tipo])
  seleccion = eh_tipo & eh_limite
  datos_seleccion = datos[seleccion]
  datos_new = pd.concat([datos_new, datos_seleccion])
```

```python
datos_new.boxplot(['Valor'], by = ['Tipo'])
```

```python
datos_new.to_csv('alquiler_residencia_sin_outliers.csv', sep = ';', index = False)
```
