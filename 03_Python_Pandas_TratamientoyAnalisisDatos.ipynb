# Selecciones y frecuencias


```python
datos = pd.read_csv('alquiler_residencial.csv', sep = ';')
datos.head(10)
```

```python
#Seleccione solamente los inmuebles clasificados con Tipo 'Departamento'
seleccion =  datos.Tipo == 'Departamento'
n1 = datos[seleccion].shape[0]
n1
```

```python
#Seleccione solamente los inmuebles clasificados con tipos 'Casa', 'Casa en condiminio' y 'Casa de villa'
seleccion =  (datos.Tipo == 'Casa')  | (datos.Tipo == 'Casa en condominio')  | (datos.Tipo == 'Casa de villa')
n2 = datos[seleccion].shape[0]
n2
```

```python
#Seleccione los inmuebles con área entre 60 y 100 metros cuadrados, incluyendo los límites
seleccion = (datos['Area'] >=60) & (datos['Area'] <=100)
n3 = datos[seleccion].shape[0]
n3
```

```python
#Selecciones los inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2.000,00
seleccion = (datos['Cuartos'] >= 4) & (datos['Valor'] < 2000)
n4 = datos[seleccion].shape[0]
n4
```

```python
print('La cantidad de Departamentos son {}'.format(n1))
print('La cantidad de casas, Casa en condominio y casa de villa es {}'.format(n2))
print('los inmuebles entre 60 y 100 metros cuadrados son {}'.format(n3))
print('Los inmuebles de al menos 4 dormitorios y con un coste de alquiler menos a $2.000,00 son {}'.format(n4))
```

Tratamiento de datos faltantes

```python
datos = pd.read_csv('alquiler_residencial.csv', sep = ';')
datos.head(10)
```

```python
#Método para reconocer nulos
#is null coloca true cuando la condicion es verdadera
datos.isnull()
```

```python
datos.notnull()
```

```python
datos.info()
```

```python
#Traer las columnas los cuales los valores son nulos:
datos[datos['Valor'].isnull()]
```

```python
A = datos.shape[0]
```

```python
#Se elimina para que no impacte en el modelo
datos.dropna(subset = ['Valor'], inplace = True)
B= datos.shape[0]
A-B #Muestra la cantidad de lineas eliminadas
```

Tratamiendo de Datos Faltantes


```python
#Si el local es mantenimiento y el mantenimiento es nulo se puede eliminar, lo cual supondría una mal carga de dato.
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
seleccion = (datos['Mantenimiento'].isnull()) & (datos['Tipo'] == 'Departamento')
```

```python
A = datos.shape[0]
datos = datos[~seleccion] # ~ Indica lo contrario
B= datos.shape[0]
A-B #Muestra la cantidad de lineas eliminadas
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
datos = datos.fillna({'Mantenimiento': 0, 'Impuesto': 0})
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
datos[datos['Impuesto'].isnull()].shape[0]
```
